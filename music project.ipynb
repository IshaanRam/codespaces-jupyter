# 2. Key detection using Essentia (simplest)

````python
import essentia.standard as es

# load audio
audio = es.MonoLoader(filename="tryc.mp3")()

# detect key
key, scale, strength = es.KeyExtractor()(audio)

print("Key:", key)
print("Scale:", scale)
print("Confidence:", strength)
````

Example output:

```
Key: D
Scale: minor
Confidence: 0.81
```

This one is the **most plug-and-play**.

---

# 3. Key detection using Librosa (chroma method)

Here the model looks at **pitch class energy** across the song.

````python
import librosa
import numpy as np

y, sr = librosa.load("tryc.mp3")

# compute chroma
chroma = librosa.feature.chroma_cqt(y=y, sr=sr)

# average pitch class strength
chroma_mean = np.mean(chroma, axis=1)

pitch_classes = ['C','C#','D','D#','E','F','F#','G','G#','A','A#','B']
key_index = np.argmax(chroma_mean)

print("Estimated key:", pitch_classes[key_index])
````

This gives the **tonic note**, but not necessarily major/minor unless you add scale templates.

Still useful.

# 4. Reality check (before you trust the machine too much)

Songs often:

* change key
* borrow chords
* use modes
* have ambiguous tonality

So the model might say:

```
C major (0.55)
```

Which basically translates to
*"probably C major but please don’t start a music theory war."*
